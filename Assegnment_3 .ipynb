```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

def  zScore(x  ,  u , s ):
    return(x-u)/s


u = 5 
s = 2
xl = 2 
xu= 8

# p(2<x<8)  --> p =p((xu-u)/s)-p((xl-u)/s) -->p(zu) - p(zl)

zl= zScore(xl , u ,s)
zu= zScore(xu , u ,s)
p=stats.norm.cdf(zu)-stats.norm.cdf(zl)
print (p)
```

```python
import statistics
dataCAt = [2.31, 2.09, 2.36, 1.95, 1.98, 2.25, 2.16, 2.07, 1.88, 1.94, 1.97, 2.02]
n = 12
mean = statistics.mean(dataCAt)
# standard deviation of data 
std = statistics.stdev(dataCAt)

# standard error of the mean
stdErrorMean = std/(n**(1/2))

#t distribution value
tScore = 2.201
# Confidance interval 
print( "[ " , mean-tScore * stdErrorMean , " , " ,mean+ tScore * stdErrorMean , " ]")

```
